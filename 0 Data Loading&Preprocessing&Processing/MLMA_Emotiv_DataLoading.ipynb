```python
import os
import pandas as pd

# Function: Fetch 8 EEG for 27 participants
def find_eeg_csv_files(root_dir, excluded_suffixes):
    eeg_csv_files = []
    # 27 participants.
    for test_num in range(27):
        # 2 phases.
        for phase in ["Training", "Testing"]:
            # 4 Activities.
            for activity in ["Activity_One", "Activity_Two", "Activity_Three", "Activity_Four"]:
                # Construct the path to the Emotiv folder.
                emotiv_dir = os.path.join(root_dir, f"Test{test_num}", phase, activity, "Emotiv")
                if os.path.exists(emotiv_dir):  # Check if the path exists
                    for file in os.listdir(emotiv_dir):
                        if file.endswith('.csv') and not file.endswith(tuple(excluded_suffixes)):
                            eeg_csv_files.append(os.path.join(emotiv_dir, file))
    return eeg_csv_files


# The path to the Emotiv-based-Data folder
root_dir = '/content/drive/MyDrive/MLMA Project/BS-HMS-Dataset/Emotiv-based-Data'

# Excluded file suffixes
excluded_suffixes = ['bp.csv', 'md.csv', 'pm.csv']

# Find all relevant EEG csv files
eeg_csv_files = find_eeg_csv_files(root_dir, excluded_suffixes)




```

```python
def load_csv_data(files_list):
    data_frames = {}
    for file_index, file_path in enumerate(files_list):
        test_num = file_index // (4 * 2)
        activity_num = (file_index % 8) + 1

        df = np.genfromtxt(file_path, delimiter=',', skip_header=1)
        df = pd.DataFrame(df)
        df.columns=[
            "COUNTER", "INTERPOLATED", "AF3", "F7", "F3", "FC5", "T7", "P7", "O1", "O2",
            "P8", "T8", "FC6", "F4", "F8", "AF4", "RAW_CQ", "GYROX", "GYROY", "MARKER",
            "MARKER_HARDWARE", "SYNC", "TIME_STAMP_s", "TIME_STAMP_ms", "TimeStamp",
            "CQ_AF3", "CQ_F7", "CQ_F3", "CQ_FC5", "CQ_T7", "CQ_P7", "CQ_O1", "CQ_O2",
            "CQ_P8", "CQ_T8", "CQ_FC6", "CQ_F4", "CQ_F8", "CQ_AF4", "CQ_CMS", "CQ_DRL"
        ]
        data_name = f'EEG_Test{test_num}_{activity_num}'
        df.to_csv(f'/content/drive/MyDrive/MLMA Project/BS-HMS-Dataset/Emotiv EEG/{data_name}.csv', index=False)

# Load all the found EEG csv files into pandas DataFrames
eeg_data_frames = load_csv_data(eeg_csv_files)
```
